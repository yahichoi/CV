```python
# 예제 11.3-1 프로그램

# camshift algorithm

import cv2
import numpy as np

# 비디오 파일 열기
cap = cv2.VideoCapture('slow.flv')

if not cap.isOpened():
    print('Video open failed!'), sys.exit()
    
ret, frame = cap.read()
if not ret:
    print('frame read failed!'), sys.exit()

# 마우스 이용하여 관심영역 선택 
#'엔터' 또는 '스페이스' 키 눌러 선택 완료
rc = cv2.selectROI("ROI Setting", frame, True)
#rc = (x,y,w,h)
    
roi = frame[rc[1]:rc[1]+rc[3], rc[0]:rc[0]+rc[2]]
roi_hsv = cv2.cvtColor(roi, cv2.COLOR_BGR2HSV)

# HS 히스토그램 계산
channels = [0, 1]
ranges = [0, 180, 0, 256]
hist=cv2.calcHist([roi_hsv], channels, None, [90, 128], ranges)

# CamShift 알고리즘 종료 기준
term_crit=(cv2.TERM_CRITERIA_EPS|cv2.TERM_CRITERIA_COUNT,10,0.6)

# 비디오 매 프레임 처리
while True:
    ret, frame = cap.read()
    
    if not ret:
        break
        
    # HS 히스토그램에 대한 역투영
    frame_hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    backproj = cv2.calcBackProject([frame_hsv], channels, 
                                   hist, ranges, 1)
    
    # CamShift
    ret, rc = cv2.CamShift(backproj, rc, term_crit)
    
    # 추적 결과 화면 출력
    cv2.rectangle(frame, rc, (0, 0, 255), 2)
    
    cv2.waitKey()
    
    cv2.imshow('Tracking Region', frame)
    
    if cv2.waitKey(60) == 27:
        break
        
cap.release()
cv2.destroyAllWindows()
```
