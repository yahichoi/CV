```python
# 예제 8.1-2 프로그램 
    
import cv2
import numpy as np

#  영상 읽어 크기 변경
image = cv2.imread('polygon.png',-1)
img = cv2.resize(image, dsize=(256,256))
img1 = img.copy()
img2 = img.copy()
src = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


### < 해리스 코너 검출 >
corners = cv2.cornerHarris(src, 3, 3, 0.04)

# 변화량 결과의 최대값 10% 이상의 좌표 구하기 
coord = np.where(corners > 0.05* corners.max())
coord = np.stack((coord[1], coord[0]), axis=-1)

# 코너 좌표에 동그리미 그리기
print('# of Harris coord =', coord.shape[0])
for x, y in coord:
    cv2.circle(img, (x, y), 5, (255,0,0),12, cv2.LINE_AA)

### < goodFeaturesToTrack 코너 검출(Shi-Tomasi) >
corners = cv2.goodFeaturesToTrack(src, 100, 0.05, 10)

# 실수 좌표를 정수 좌표로 변환
corners = np.int32(corners)

# 좌표에 동그라미 표시
print('# of GFTT corners =', corners.shape[0])
for corner in corners:
    x, y = corner[0]
    cv2.circle(img1, (x, y), 5, (255,0,0), 1, cv2.LINE_AA)
   
### < FAST코너 검출 >
fast = cv2.FastFeatureDetector_create(30)

# cv2.KeyPoint
keypoints = fast.detect(src)

# 좌표에 동그라미 표시
print('# of FAST corners =', len(keypoints))
for kp in keypoints:
    pt = (int(kp.pt[0]), int(kp.pt[1]))
    cv2.circle(img2, pt, 5, (255,0,0), 1, cv2.LINE_AA)

images = [img,img1,img2]
titles = ['Harris','GFTT','FAST']

plt.figure(figsize=(12,4))
for i in range(3) :
    plt.subplot(1,3,i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
plt.show()
  
```
